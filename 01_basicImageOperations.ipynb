## Basic image operations

```python
#import required modeules
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

#### Reading, displaying and writing an image

```python
#Reading an image
img = cv2.imread('g_and_a.jpeg')
```

```python
# displyaing an image
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#wirting an image to file
cv2.imwrite('messigray.png',img)
```

```python
# simple code combining all the above 
cv2.imshow('image',img)
k = cv2.waitKey(0)
if k == 27:         # wait for ESC key to exit
    cv2.destroyAllWindows()
elif k == ord('s'): # wait for 's' key to save and exit
    cv2.imwrite('messigray.png',img)
    cv2.destroyAllWindows()
```

```python
#printing using matplot lib (note that the channel are reversed)
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

#### Accessing and Modifying pixel values

```python
#accessing a pixel BGR
px = img[100,100]
print(px)
```

```python
#Accessing pixel element of a channel
blue = img[100,100,0]
green = img[100,100,1]
red = img[100,100,2]
print('Blue pixel value:',blue,'Green pixel value:',green,'Red pixel value:',red)
```

```python
# modifying pixel value
img[100,100] = [255,255,255]
print (img[100,100])
```

```python
# accessing RED channel pixel value
img.item(10,10,2)
```

```python
# modifying RED channel pixel value
img.itemset((10,10,2),100)
img.item(10,10,2)
```

#### Accessing Image Properties

```python
#image shape
print(img.shape)
```

```python
#image size (totoal number of pixels)
print(img.size)
```

```python
# image datatype
print (img.dtype)
```

```python
# copying a group of pixels
glasses = img[200:280, 370:560,:]
img2 = img
img2[173:253, 90:280] = glasses
cv2.imshow('image',img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Splitting and merging an image channels


```python
b,g,r = cv2.split(img)
img = cv2.merge((b,g,r))
```

```python
plt.imshow(b)
```

```python
b2 = img[:,:,0]
plt.imshow(b2)
cv2.imshow('image',b2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Image borders

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

BLUE = [255,0,0]

img1 = cv2.imread('g_and_a.jpeg')

replicate = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REPLICATE)
reflect = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REFLECT)
reflect101 = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REFLECT_101)
wrap = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_WRAP)
constant= cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_CONSTANT,value=BLUE)

plt.subplot(231),plt.imshow(img1,'gray'),plt.title('ORIGINAL')
plt.subplot(232),plt.imshow(replicate,'gray'),plt.title('REPLICATE')
plt.subplot(233),plt.imshow(reflect,'gray'),plt.title('REFLECT')
plt.subplot(234),plt.imshow(reflect101,'gray'),plt.title('REFLECT_101')
plt.subplot(235),plt.imshow(wrap,'gray'),plt.title('WRAP')
plt.subplot(236),plt.imshow(constant,'gray'),plt.title('CONSTANT')

plt.show()
```
